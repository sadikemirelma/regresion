```python
df=pd.read_csv("/kaggle/input/audi-a1-listings/Audi_A1_listings.csv")
```

```python
df=df.drop(columns=['href','index','MileageRank','PriceRank','PPYRank','Score'])
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat","ppy"]
df.info()
```

```python
df.head(3)
```

```python
df.info()
```

```python
df["motor"]=df["motor"].str.replace("L","")
df["motor"]=pd.to_numeric(df["motor"])
```

```python

df.info()
```

```python
df=pd.get_dummies(df,columns=["kasa","vites","yakit"],drop_first=True)
```

```python
df.info()
```

```python
y=df[["fiyat"]]
x=df.drop("fiyat",axis=1)
```

```python
# lm=LinearRegression()
# model=lm.fit(x,y)
# model.score(x,y)
# model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=22)
```

```python
lm=LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
