```python
df=pd.read_csv('/kaggle/input/student-marks-dataset/Student_Marks.csv')
```

```python
df.columns=['sinif','saat','puan']
df.head(3)
```

```python
y= df[['puan']]
x= df[['sinif','saat']]
```

```python
l= LinearRegression()
model=l.fit(x,y)
```

```python
model.predict([[3,2]])
```
